```python
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv", index_col="PassengerId")
print(train.shape)
train.head()
```

```python
test = pd.read_csv("data/test.csv", index_col="PassengerId")
print(test.shape)
test.head()
```

```python
gendersub = pd.read_csv("data/gender_submission.csv", index_col="PassengerId")
print(gendersub.shape)
gendersub.head()
```

## Preprocessing
## Encode Sex

```python
train.loc[train["Sex"] == "male", "Sex_encode"] = 0
train.loc[train["Sex"] == "female", "Sex_encode"] = 1

print(train.shape)
train[["Sex", "Sex_encode"]].head()
```

```python
test.loc[test["Sex"] == "male", "Sex_encode"] = 0
test.loc[test["Sex"] == "female", "Sex_encode"] = 1

print(test.shape)
test[["Sex", "Sex_encode"]].head()
```

## Fill in missing fare

```python
train["Fare_fillin"] = train["Fare"]
test["Fare_fillin"] = test["Fare"]

print(test.shape)
test[["Fare", "Fare_fillin"]].head()
```

```python
test.loc[test["Fare"].isnull(), "Fare_fillin"] = 0
test.head()
```

## Encode Embarked
## One Hot Encoding



```python
# C == [1, 0, 0]
# S == [0, 1, 0]
# Q == [0, 0, 1]

# C == [True, False, False]
# C == [False, True, False]
# C == [False, False, True]
```

```python
train["Embarked_C"] = train ["Embarked"] == "C"
train["Embarked_S"] = train ["Embarked"] == "S"
train["Embarked_Q"] = train ["Embarked"] == "Q"

print(train.shape)
train[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
test["Embarked_C"] = test ["Embarked"] == "C"
test["Embarked_S"] = test ["Embarked"] == "S"
test["Embarked_Q"] = test ["Embarked"] == "Q"

print(test.shape)
test[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
# Feature / Label
# Pclass / Survived
# Sex
# Fare
# Embarked
```

```python
feature_names = ["Pclass", "Sex_encode", "Fare_fillin", "Embarked_C", "Embarked_S", "Embarked_Q"]
feature_names
```

```python
X_train = train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
label_name = "Survived"

y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

## Decision Tree

```python
# scikit-learn == sklearn
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=5)
model
```

```python
# 1) fit / 2) predicit
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions.shape)
predictions[0:10]
```

## submit

```python
submit = pd.read_csv("data/gender_submission.csv", index_col="PassengerId")

print(submit.shape)
submit.head()
```

```python
submit["Survived"] = predictions

print(submit.shape)
submit.head()
```

```python
submit.to_csv("data/decision-tree.csv")
```
